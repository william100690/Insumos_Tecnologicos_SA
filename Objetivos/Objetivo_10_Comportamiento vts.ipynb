```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#leer el archivo csv
df = pd.read_csv("C:/Users/willi/OneDrive/Escritorio/Proyectos/Insumos_Tecnologicos_SA/Datasets/Venta.csv")
```

```python
#mostrar archivo csv
df.head(5)
```

```python
#verificar el tipo de dato de cada columna
df.dtypes
```

```python
# cambiar el tipo de dato a tipo date time
df['Fecha'] = pd.to_datetime(df['Fecha'])

df.dtypes
```

```python
#obtener únicamente los años
años = df['Fecha'].dt.year.unique()
años.sort()
años
```

```python
#ordenar el data frame por fecha
df = df.sort_values('Fecha')
df.head(3)
```

```python
#obtener las ventas al multiplicar la columna precio y cantidad
df['Ventas'] = df['Precio'] * df['Cantidad']
df.head(3)
```

```python
#crear un nuevo data frame con la suma de las ventas de cada fecha
df_Fecha_Ventas = df.groupby('Fecha')['Ventas'].sum().reset_index()
```

```python
df_Fecha_Ventas
```

```python
#obtener un data frame por cada año
df_2015 = df_Fecha_Ventas[df_Fecha_Ventas['Fecha'].dt.year == 2015].copy()
df_2016 = df_Fecha_Ventas[df_Fecha_Ventas['Fecha'].dt.year == 2016].copy()
df_2017 = df_Fecha_Ventas[df_Fecha_Ventas['Fecha'].dt.year == 2017].copy()
df_2018 = df_Fecha_Ventas[df_Fecha_Ventas['Fecha'].dt.year == 2018].copy()
df_2019 = df_Fecha_Ventas[df_Fecha_Ventas['Fecha'].dt.year == 2019].copy()
df_2020 = df_Fecha_Ventas[df_Fecha_Ventas['Fecha'].dt.year == 2020].copy()
```

```python
#obtner la grafica de comportamiento de las ventas en el año 2015
plt.plot(df_2015['Fecha'], df_2015['Ventas'])

# Configurar el eje X para mostrar las fechas de manera legible
plt.xticks(rotation=45)

# Etiquetas de los ejes
plt.xlabel('Fecha')
plt.ylabel('Ventas')

# Título del gráfico
plt.title('Ventas por Fecha')

# Mostrar el gráfico
plt.show()
```
